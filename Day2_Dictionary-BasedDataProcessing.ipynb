딕셔너리 활용 예시

```python
#각 학생의 과목 평균 점수 구하기
def avg(scores):
  return sum(scores) / len(scores)


students = {
    "Alice": [85, 90, 95],
    "Bob": [70, 80, 65],
    "Charlie": [100, 100, 100],
}


for name, scores in students.items():
  average = avg(scores)
  print(f"{name}의 평균 점수: {average}")
```

```python
#과목 평균 점수가 90점 이상인 학생 구하기
for name, scores in students.items():
  average = avg(scores)
  if average >= 90:
    print(f"{name}은 우수 학생입니다! 평균:{average}")
```

```python
#평균 점수가 가장 높은 학생 구하기
best_stduent = ""
best_avg = 0

for name, scores in students.items():
  average = avg(scores)
  if average > best_avg:
    best_student = name
    best_avg = average

print(f"최고 점수 학생: {best_student}, 평균: {best_avg}")
```

json 파일을 파싱하여 데이터 분석

```python
import pandas as pd

csv_path = "/content/drive/MyDrive/Colab Notebooks/2025/students_scores.csv" #구글 드라이브 주소

df = pd.read_csv(csv_path, index_col=0)

df.to_json("/content/drive/MyDrive/Colab Notebooks/2025/students_scores.json",
           orient="index",
           force_ascii=False,
           indent=2)
```

```python
import json

with open("/content/drive/MyDrive/Colab Notebooks/2025/students_scores.json", "r", encoding="utf-8") as f:
    data = json.load(f)

print(list(data.items())[:3])  # 앞부분만 보기
```

```python
#Student004의 전체 성적 출력
print(data["Student004"])
```

```python
#앞 10명의 korean 성적 출력
for student, scores in list(data.items())[:10]:
  for subject , score in scores.items():
    if subject == 'korean':
      print(f"{student}의 {subject} 성적은 : {score}입니다.")
```

```python
#전체 학생의 영어 평균 점수 출력
total = 0
count = 0
for student, scores in data.items():
  for subject, score in scores.items():
    if subject == 'english':
        total += score
        count += 1
sub_avg = total / count

print(f"전체 학생의 영어 평균 점수는 {sub_avg:.2f}점 입니다.")
```

```python
#특정 학생의 전체 과목 평균 점수
scores = data["Student120"]
stu_avg = sum(scores.values()) / len(scores)

print(f"Student120의 전체 과목 평균 점수는 {stu_avg:.2f}점입니다")
```

```python
#가장 평균 점수가 높은 학생 출력
best_student = ""
best_avg = 0

for student , scores in data.items():
  total = sum(scores.values())
  count = len(scores)
  avg = total / count

  if avg > best_avg:
    best_student = student
    best_avg = avg

print(f" 가장 평균 점수가 높은 학생: {best_student}, 평균: {best_avg:.2f}점")
for subject, score in data[best_student].items():
  print(f"{subject}: {score}")
```

```python
#가장 평균 점수가 낮은 학생 출력
worst_student = ""
worst_avg = 100

for student , scores in data.items():
  total = sum(scores.values())
  count = len(scores)
  avg = total / count

  if avg < worst_avg:
    worst_student = student
    worst_avg = avg

print(f" 가장 평균 점수가 낮은 학생: {worst_student}, 평균: {worst_avg:.2f}점")
for subject, score in data[worst_student].items():
  print(f"{subject}: {score}")
```

```python
#전체 학생의 영어 점수를 상위 10명까지 내림차순으로 정렬
sorted_english = sorted(data.items(), key=lambda item: item[1]["english"], reverse=True)

print("영어 성적 기준 상위 10명:")
for student, scores in sorted_english[:10]:
    print(f"{student}: {scores['english']}점")
```

```python
#전체 학생의 과목 평균 점수를 내림차순으로 정렬
sorted_by_avg = sorted(
    data.items(),
    key=lambda item: sum(item[1].values()) / len(item[1]),
    reverse=True
)

for student, scores in sorted_by_avg:
    avg = sum(scores.values()) / len(scores)
    print(f"{student}: 평균 {avg:.2f}점")
```

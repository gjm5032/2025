소수는 1보다 큰 자연수 중 1과 그 자신의 수로만 나누어 떨어지는 수를 말한다.

파이썬으로 소수를 출력하는 코드를 짠다.

1000까지 소수의 값 구하기

```python
#처음 짠 코드
for i in range(2,1000):
  for j in range(2,i): # i가 2일 땐 range(2,2)이기 때문에 j 루프가 실행되지 않음
    if i % j == 0: #나누어 떨어지면 소수 아님
      break
  else: # 이 else는 for j in range(2, i): 와 연결되어 있음 , break 없이 루프가 정상 종료되면 소수 else: 실행
      print(i , end= " ")
```

```python
#for문을 while 문으로 바꾸기 while - else 방식

i = 2
while i < 1000:
  j = 2
  while j < i:
    if i % j == 0:
      break
    j += 1
  else :
    print(i , end= " ")
  i += 1
```

```python
#for문을 while 문으로 바꾸기 flag 방식
i = 2
while i < 1000:
  j = 2
  prime = True # prime 변수를 True로 설정
  while j < i:
    if i%j==0: # 만약 i가 j로 나누어 떨어진다면
      prime = False # pirme은 False
      break
    j += 1

  if prime: #prime이 True면
    print(i , end= " ") #소수 출력
  i += 1
```

에라토스테네스의 체

고대 그리스 수학자 에라토스테네스가 고안해낸 소수를 찾는 방법이다.

예를들어 100까지의 소수를 찾는다 가정했을 때

먼저 2를 제외한 2의 배수를 모두 지운다.
그다음 3을 제외한 3의 배수를 지운다.
4는 이미 2의 배수이므로 생략한다.
5를 제외한 5의 배수를 지운다.
...

이 과정을 계속 반복하면, 남는 수는 모두 소수다.

이러한 방법은 정해진 범위 내에서 모든 소수를 한 번에 구할 때 빠르고 효율적인 방법 중 하나이다.

```python
N = 1000
is_prime = [True] * (N + 1)  # 0부터 N까지 총 N+1개의 True 값을 갖는 리스트 생성

is_prime[0] = is_prime[1] = False  # 0과 1은 소수가 아니므로 False로 설정

for i in range(2, N):
    if is_prime[i]:  # i가 아직 소수로 남아있다면
        # i의 배수들을 지움 (i*i부터 시작, 그 이하 배수는 이미 처리됨)
        for j in range(i * i, N + 1, i):
            is_prime[j] = False

# True로 남아 있는 인덱스(소수) 출력
for i in range(2, N + 1):
    if is_prime[i]:
        print(i, end=" ")
```
